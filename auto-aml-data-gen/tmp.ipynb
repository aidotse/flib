```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = '/home/edvin/Desktop/flib/AMLsim/outputs/tmp/tx_log.csv'
df = pd.read_csv(file)
df
```

```python
from preprocess import preprocess

dfs = preprocess(path_to_tx_log=file, banks=['bank'], split_type='spatial')
df_train, df_test = dfs[0]
df_train
```

```python
df_neg = df_train[df_train['is_sar'] == 0]
df_pos = df_train[df_train['is_sar'] == 1]

columns = df_train.columns[2:-1]
#columns = ['counts_in_0_29', 'counts_out_0_29']

n_subplots = len(columns)
n_rows = 3 #int(np.sqrt(n_subplots))
n_cols = 7 #int(np.ceil(n_subplots/n_rows))
fig, axs = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 5*n_rows))

for ax, column in zip(axs.flatten(), columns):
    ax.hist(df_neg[column], bins=50, alpha=0.5, label='Non-SAR')
    ax.hist(df_pos[column], bins=50, alpha=0.5, label='SAR')
    #x_min = df_train[column].min()
    #x_max = df_train[column].max()
    #ax.set_xlim(x_min-10, x_max+10)
    ax.set_title(column)
    ax.set_yscale('log')
    ax.legend()

plt.tight_layout()
plt.show()
```
