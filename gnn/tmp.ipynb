```python
import pandas as pd
```

```python
path = '/home/edvin/Desktop/flib/AMLsim/outputs/10K_accts/tx_log.csv'
df_raw = pd.read_csv(path)
```

```python
df_raw.head()
```

```python
df = df_raw[(df_raw['bankOrig']!='source') & (df_raw['bankDest']!='sink')]
```

```python
banks = ['bank']
for bank in banks:
    df_bank = df[(df['bankOrig']==bank) | (df['bankDest']==bank)]
    print(f'n txs within/to/from {bank}: {len(df_bank)}')
    n_sar = len(df_bank[df_bank['isSAR']==1])
    print(f'  n sar txs: {n_sar}')
    n_non_sar = len(df_bank[df_bank['isSAR']==0])
    print(f'  n non-sar txs: {n_non_sar}')
    print(f'  ratio sar/non-sar: {n_sar/n_non_sar}')
```
